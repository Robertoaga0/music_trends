```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from datetime import datetime, timedelta
```

```python


# Crear una lista de fechas semanales desde 2010 hasta hoy
start_date = datetime(2010, 1, 1)  # Desde 1 de enero de 2010
end_date = datetime.today()  # Hasta hoy
dates = []

while start_date <= end_date:
    dates.append(start_date.strftime("%Y-%m-%d"))  # Formato YYYY-MM-DD
    start_date += timedelta(days=7)  # Avanzamos una semana

# Lista para almacenar los datos
data = []

# Simular un navegador
headers = {"User-Agent": "Mozilla/5.0"}

# Recorrer cada fecha y obtener los datos
for date in dates:
    url = f"https://www.billboard.com/charts/hot-100/{date}/"
    print(f"Scrapeando datos de {date}...")  # Para ver el progreso

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Extraer datos
    songs = []
    artists = []
    positions = []

    for i, entry in enumerate(soup.select("li.o-chart-results-list__item")):
        song = entry.select_one("h3#title-of-a-story")
        artist = entry.select_one("span.a-no-trucate")
        
        if song and artist:
            songs.append(song.text.strip())
            artists.append(artist.text.strip())
            positions.append(i + 1)

    # Guardar los datos con la fecha correspondiente
    for pos, song, artist in zip(positions, songs, artists):
        data.append([date, pos, song, artist])

    # Esperar un poco para evitar bloqueos del sitio
    time.sleep(1.5)

# Crear un DataFrame
df = pd.DataFrame(data, columns=["Fecha", "Posición", "Canción", "Artista"])

# Guardar en CSV
df.to_csv("billboard_hot100_semanal.csv", index=False)

print("Scraping completado. Datos guardados en 'billboard_hot100_semanal.csv'")
```
